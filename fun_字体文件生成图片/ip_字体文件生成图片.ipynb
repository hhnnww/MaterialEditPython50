```python
import math
from pathlib import Path

from PIL import Image, ImageDraw, ImageFont

from fun_字体文件生成图片.fun_获取一个成语API import Class_随机获取一个成语
from MaterialEdit.fun_图片编辑.fun_删除图片边框.fun_删除图片边框 import fun_删除图片边框
from MaterialEdit.fun_图片编辑.fun_图片扩大粘贴 import fun_图片扩大粘贴
from MaterialEdit.fun_图片编辑.fun_图片拼接.fun_图片竖向拼接 import fun_图片竖向拼接

ma_path = Path(r"F:\小夕素材\5000-5999\5168")
text_color = (255, 255, 255, 255)
bg_color = (0, 0, 0, 255)

for font_path in ma_path.rglob("*"):
    if font_path.suffix.lower() not in [".ttf", ".otf", ".ttc"]:
        continue

    img_path = font_path.with_suffix(".png")

    if img_path.exists():
        continue

    text = Class_随机获取一个成语.fun_成语文件()

    # 写中文
    true_font = ImageFont.truetype(font=font_path.as_posix(), size=500)
    size = true_font.getbbox(text=text.word)
    bg = Image.new(
        "RGBA",
        (math.ceil(size[2] + size[0]), math.ceil(size[3] + size[1])),
        bg_color,
    )
    draw = ImageDraw.Draw(bg)
    draw.text((0, 0), text.word, fill=text_color, font=true_font)
    bg = fun_删除图片边框(bg)

    # 写拼音
    py_path = img_path.with_stem(f"{img_path.stem}_py")
    true_font = ImageFont.truetype(font=font_path.as_posix(), size=500)
    size = true_font.getbbox(text=text.py)
    py_bg = Image.new(
        "RGBA",
        (math.ceil(size[2] + size[0]), math.ceil(size[3] + size[1])),
        bg_color,
    )
    draw = ImageDraw.Draw(py_bg)
    draw.text((40, 0), text.py, fill=text_color, font=true_font)

    py_bg.thumbnail((int(bg.width / 2), 999999), resample=Image.Resampling.LANCZOS)
    py_bg = fun_删除图片边框(py_bg)

    bg = fun_图片竖向拼接(
        image_list=[bg, py_bg],
        spacing=20,
        align_item="center",
        background_color=bg_color,
    )

    width = int(bg.width * 1.3)
    height = int(((bg.width * 1.3) / 4) * 2)
    if height < bg.height:
        height = int(bg.height * 1.3)

    bg = fun_图片扩大粘贴(
        im=bg,
        width=width,
        height=height,
        left="center",
        top="center",
        background_color=bg_color,
    )

    bg.save(img_path.as_posix())
```
