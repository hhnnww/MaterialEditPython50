```python
from pathlib import Path
from pprint import pprint

from PIL import Image
import random
from MaterialEdit.fun_图片编辑 import ImageEdit
from MaterialEdit.fun_制作首图.style_黑鲸 import fun_黑鲸首图

gutter = 10
line_col = 5
line_row = 2

small_width = int((1500 - ((line_col + 1) * gutter)) / line_col)
small_height = int(((1300) - ((line_row + 1) * gutter)) / line_row)

comb_image_list = []
in_image_list = []

path_list = []
for in_file in Path(r"E:\\小夕素材\\10000-20000\\10280\\预览图").rglob("*"):
    if in_file.is_file():
        path_list.append(in_file)

random.shuffle(path_list)
while len(path_list) < (line_col + 1) * line_row:
    path_list += path_list

for in_file in path_list:
    im = Image.open(in_file.as_posix()).convert("RGBA")
    im = ImageEdit.fun_图片裁剪(
        im=im, width=small_width, height=small_height, position="start"
    )
    # im = ImageEdit.fun_图片画边框(im=im, border_color=(230, 230, 250, 255))
    # im = ImageEdit.fun_图片切换到圆角(
    #     im=im, border_radius=8, background_color=(255, 255, 255, 255)
    # )
    in_image_list.append(im)

    if len(in_image_list) == line_col + 1:
        comb_image_list.append(in_image_list.copy())
        in_image_list = []

    if len(comb_image_list) == line_row:
        break

new_comb_im = []
for num, line_comb in enumerate(comb_image_list):
    line_im: Image.Image = ImageEdit.fun_图片横向拼接(
        image_list=line_comb,
        spacing=10,
        align_item="center",
        background_color=(255, 255, 255, 255),
    )
    left = int((line_im.width - 1500) / 2)
    line_im = line_im.crop((left, 0, left + 1500, line_im.height))
    new_comb_im.append(line_im)

bg = ImageEdit.fun_图片竖向拼接(
    image_list=new_comb_im,
    spacing=10,
    align_item="center",
    background_color=(255, 255, 255, 255),
)

newbg: Image.Image = ImageEdit.fun_图片扩大粘贴(
    im=bg,
    width=1500,
    height=1300,
    left="center",
    top="center",
    background_color=(255, 255, 255, 255),
)

shadow_im = Image.open(r"C:\\Users\\wuweihua\\Desktop\\UPLOAD\\shadow.png")
newbg.paste(shadow_im, (0, 0), shadow_im)

im = fun_黑鲸首图(
    im=newbg, title="33套 清明节国风海报", material_format="psd", material_id="10387"
)
im = im.convert("RGB")
im.save(r"C:\Users\wuweihua\Desktop\UPLOAD\st1.jpg")
```
