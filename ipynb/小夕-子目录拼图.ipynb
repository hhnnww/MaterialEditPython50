```python
import math
from pathlib import Path

from PIL import Image

from MaterialEdit.fun_图片编辑 import fun_图片扩大粘贴, fun_图片横向拼接
from MaterialEdit.fun_遍历图片 import fun_遍历图片


def fun_处理单个文件夹(in_path: Path, effect_path: Path):
    image_list = []

    for image in fun_遍历图片(
        folder=in_path.as_posix(),
        used_image_number=0,
        image_sort=True,
    ):
        im = Image.open(image)
        image_list.append(im)

    image_list.sort(key=lambda k: k.width / k.height, reverse=True)

    max_height = max([im.height for im in image_list])

    for num, im in enumerate(image_list):
        width = math.ceil(max_height * (im.width / im.height))
        im = im.resize((width, max_height), resample=Image.Resampling.LANCZOS)
        image_list[num] = im

    bg = fun_图片横向拼接(
        image_list=image_list,
        spacing=10,
        align_item="start",
        background_color=(255, 255, 255, 255),
    )

    bg = fun_图片扩大粘贴(
        im=bg,
        width=bg.width + 20,
        height=bg.height + 20,
        left="center",
        top="center",
        background_color=(255, 255, 255, 255),
    )

    if effect_path.exists() is not True:
        effect_path.mkdir(parents=True)

    bg.save(rf"{effect_path}\{in_path.stem}.png")

    return bg


for in_path in Path(r"F:\小夕素材\10000-20000\10667\10667").iterdir():
    if in_path.is_dir():
        fun_处理单个文件夹(
            in_path=in_path, effect_path=Path(r"F:\小夕素材\10000-20000\10667\效果图")
        )
```
