```python
from pathlib import Path

from win32com.client import Dispatch

app = Dispatch("Illustrator.Application")
```

```python
def fun_convert_ai(ai_file: str):
    app.Open(ai_file)
    doc = app.ActiveDocument

    Arguments = []
    app.DoJavaScriptFile(
        r"C:\Program Files\Adobe\Adobe Illustrator 2023\Presets\zh_CN\脚本\1-导出图片.jsx",
        Arguments,
        1,
    )
    doc.Close(2)


start_stem = 3544

for in_path in Path(r"F:\饭桶设计\3000-3999").iterdir():
    if in_path.is_dir() and int(in_path.stem) >= 3544:
        all_jpg = []

        for in_jpg_file in in_path.rglob("*"):
            if in_jpg_file.is_file() and in_jpg_file.suffix.lower() in [".jpg", ".png"]:
                all_jpg.append(in_jpg_file.name)

        for in_file in in_path.rglob("*"):
            if in_file.is_file() and in_file.suffix.lower() in [".ai"]:
                jpg_state = False
                for jpg_stem in all_jpg:
                    if in_file.stem in jpg_stem:
                        jpg_state = True

                if jpg_state is False:
                    print(in_file.as_posix())
                    fun_convert_ai(in_file.as_posix())
```
