```python
from pathlib import Path

for in_path in Path(r'E:\泡泡素材\0000-0999').iterdir():
    if in_path.is_dir():
        if len(in_path.stem) < 4:
            print(in_path)

            new_path = in_path.with_stem(f'0{in_path.stem}')
            while len(new_path.stem) < 4:
                new_path = new_path.with_stem(f'0{new_path.stem}')

            print(new_path)

            print('\n')

            in_path.rename(new_path)
```
