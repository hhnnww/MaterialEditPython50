```python
from pathlib import Path
from PIL import Image, ImageFile

jpg_list: [ImageFile, ...] = []

folder_path = Path(r"F:\小夕素材\11000-11999\11726\11726\小夕素材(02)")
save_path = folder_path.parent / f"{folder_path}.jpg"

for infile in folder_path.iterdir():
    if infile.suffix.lower() in [".jpg"]:
        im = Image.open(infile)
        jpg_list.append(im)

bg_width = min([w.width for w in jpg_list])
bg_height = sum([w.height for w in jpg_list])
bg = Image.new("RGB", [bg_width, bg_height])

x, y = 0, 0
for in_im in jpg_list:
    in_im.thumbnail((bg_width, 9999))
    bg.paste(in_im, (x, y))
    y += in_im.height

bg.save(save_path)
```
