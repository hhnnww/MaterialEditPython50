```python
import re

import pyperclip
from requests_html import HTML, Element

from MaterialEdit.fun_素材下载.fun_插入素材 import fun_插入素材
from MaterialEdit.fun_素材下载.model_素材格式 import MaterialModel

text = pyperclip.paste()
html_text = re.findall(r'jsonp.*?\("([\s\S]*?)"\)', text)[0]
html_text = re.sub(r"\\", "", html_text)


html = HTML(html=html_text)

links = html.find(".photo a")

if type(links) is list:
    for link in links:
        link: Element
        url = list(link.absolute_links)[0]
        print(url)

        img = link.find("img", first=True)

        if img is Element:
            img = img.attrs.get("src")
            img = "https:" + img.replace("_240x240.jpg", "")
            print(img)

            mod = MaterialModel(url=url, img=img, state=False)
            fun_插入素材(
                shop_name="饭桶设计", material_site="黑鲸素材", material_model=mod
            )
```
