```python
from pathlib import Path

import pyautogui
import pyperclip



from ipynb.fun_windows_窗口操作 import fun_查找图片坐标


pyautogui.PAUSE = 1



text_file = Path.home() / "Desktop" / "output.txt"


text_file.write_text("")


with open(text_file.as_posix(), "a+") as f:


    # 最大数需要加1

    for x in range(286, 295 + 1):

        material_id = str(x)

        while len(material_id) < 4:

            material_id = "0" + material_id


        pyautogui.click(fun_查找图片坐标("search"))

        pyautogui.click(2517, 173)


        pyautogui.write(material_id)

        pyautogui.hotkey("enter")

        pyautogui.sleep(2)


        # 点击栏目


        pyautogui.click(2657, 340)


        # 点击分享


        pyautogui.click(2495, 167)

        pyautogui.sleep(1)


        pyautogui.click(fun_查找图片坐标("yongjiu"))

        pyautogui.click(fun_查找图片坐标("chuangjian"))

        pyautogui.click(fun_查找图片坐标("guanbi"))

        pyautogui.click(fun_查找图片坐标("close"))


        text = pyperclip.paste()

        text = (
            material_id
            + "\t"
            + '"'
            + "\n".join(text.split("\n")[0:-2])
            + '"\n\n'
        )


        text = text.replace(" ", "")
        print(text)


        f.write(text)
```
