```python
from pathlib import Path

from PIL import Image
from win32com.client import Dispatch

from MaterialEdit.setting import IMAGE_SUFFIX

app = Dispatch("Photoshop.Application")
material_path = Path(r"F:\饭桶设计\4000-4999\4091\4091")
for in_file in material_path.iterdir():
    if in_file.is_file() and in_file.suffix.lower() in IMAGE_SUFFIX:
        im = Image.open(in_file.as_posix())
        if im.mode == "CMYK":
            im.close()
            app.Open(in_file.as_posix())
            doc = app.ActiveDocument
            doc.ChangeMode(2)
            doc.Save()
            doc.Close(2)
```
