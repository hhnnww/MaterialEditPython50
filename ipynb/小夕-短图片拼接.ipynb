```python
import shutil
from collections.abc import Iterator
from pathlib import Path

from PIL import Image

from MaterialEdit.fun_图片编辑.fun_图片拼接.fun_图片竖向拼接 import fun_图片竖向拼接
from MaterialEdit.setting import IMAGE_SUFFIX

ma_path = Path(r"F:\小夕素材\10000-20000\10933\10933")


class OneImage:
    def __init__(self, image_path: Path) -> None:
        print(image_path)
        self.image_path = image_path

    @property
    def num(self) -> int:
        return int(self.image_path.stem.split("_")[1])

    @property
    def pil(self) -> Image.Image:
        im = Image.open(self.image_path)
        im = im.convert("RGBA")
        return im


class MaterialPathObj:
    def __init__(self, ma_path: Path) -> None:
        self.ma_path = ma_path

    @property
    def image_path(self) -> Path:
        image_path = self.ma_path / "images"
        return image_path

    @property
    def all_image(self) -> Iterator[OneImage]:
        if self.image_path.exists() is True:
            for in_path in self.image_path.iterdir():
                if in_path.is_file() and in_path.suffix.lower() in IMAGE_SUFFIX:
                    yield OneImage(image_path=in_path)

    @property
    def material_png_path(self) -> Path:
        return self.ma_path / (self.ma_path.stem + ".png")

    def main(self):
        if self.image_path.exists() is True:
            image_list = list(self.all_image)
            image_list.sort(key=lambda k: k.num, reverse=False)
            im = fun_图片竖向拼接(
                image_list=[image.pil for image in image_list],
                spacing=0,
                align_item="center",
                background_color=(255, 255, 255, 255),
            )
            im.save(self.material_png_path.as_posix())

            shutil.rmtree(self.image_path)


for ma_path in Path(r"F:\小夕素材\10000-20000\10933\10933").iterdir():
    if ma_path.is_dir():
        mobj = MaterialPathObj(ma_path=ma_path)
        mobj.main()
```
