```python
from pptx import Presentation
from pptx.util import Inches
from pathlib import Path
```

```python
for in_file in Path(r"E:\小夕素材\10000-20000\10381\10381").iterdir():
    if in_file.is_file() and in_file.suffix.lower() in [".ppt", ".pptx"]:

        ppt = Presentation(in_file.as_posix())

        for slide in ppt.slides:
            for shape in slide.shapes:
                if shape.has_text_frame:
                    for paragraph in shape.text_frame.paragraphs:
                        for run in paragraph.runs:
                            if "唐峰" in run.text:
                                run.text = run.text.replace("唐峰", "小夕")

        ppt.save(in_file.as_posix())
```
